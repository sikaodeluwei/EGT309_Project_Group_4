Converting .db (SQL Lite) to .csv

```python
import sqlite3
import pandas as pd

# 1. Connect to the database
conn = sqlite3.connect('gas_monitoring.db')

# 2. Read the specific table into a DataFrame
# Replace 'your_table_name' with the actual table name inside the .db file
df = pd.read_sql_query("SELECT * FROM gas_monitoring", conn)

# 3. Export the DataFrame to a CSV file
df.to_csv('gas_monitoring.csv', index=False)

# 4. Close the connection
conn.close()
```
